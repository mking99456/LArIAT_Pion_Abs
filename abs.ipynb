```python
#This code is designed to reproduce the plot in Ashery 

#Data from Ashery, Navon, Nakai

#Ashery: Ashery, D. Navon, I., Nzuelos, G., Walter, H. K., Pfeiffer. H. J., et al. Phys. Rev. C 22: 2173 (1981)
#Navon: 
#Nakai: Nakai, K., Kobayashi, T., Numao, T., Shibata, T. A., Chiba, J., et al. Phys. Rev. Lett. 44: 1446 (1980)
#ProtoDUNE
#LADS

#Neglecting Jones and McKeown data for now.

```

```python
# Read in data files for different probes and targets and plot them
import numpy as np
import matplotlib.pyplot as plt

#Read in the data from subfolders
#pipA_abs

#Specifically the Ashery data on pi plus absorption on the following nuclei:
#Al, Bi, C, Fe, Li, Nb 

pip_al_ashery = np.loadtxt('pipA_abs/ashery-pip-al-abs.dat', skiprows=1)
pip_bi_ashery = np.loadtxt('pipA_abs/ashery-pip-bi-abs.dat', skiprows=1)
pip_c_ashery = np.loadtxt('pipA_abs/ashery-pip-c-abs.dat', skiprows=1)
pip_fe_ashery = np.loadtxt('pipA_abs/ashery-pip-fe-abs.dat', skiprows=1)
pip_li_ashery = np.loadtxt('pipA_abs/ashery-pip-li-abs.dat', skiprows=1)
pip_nb_ashery = np.loadtxt('pipA_abs/ashery-pip-nb-abs.dat', skiprows=1)

## Load the Nakai data for al, au, cu, sn, ti

pip_al_nakai = np.loadtxt('pipA_abs/nakai-pip-al-abs.dat', skiprows=1)
pip_au_nakai = np.loadtxt('pipA_abs/nakai-pip-au-abs.dat', skiprows=1)
pip_cu_nakai = np.loadtxt('pipA_abs/nakai-pip-cu-abs.dat', skiprows=1)
pip_sn_nakai = np.loadtxt('pipA_abs/nakai-pip-sn-abs.dat', skiprows=1)
pip_ti_nakai = np.loadtxt('pipA_abs/nakai-pip-ti-abs.dat', skiprows=1)

#Load the ProtoDUNE data for Ar
pip_ar_protodune = np.loadtxt('pipA_abs/protodune-pip-ar-abs.dat', skiprows=1)

#Load the LADS data for Ar, N, Xe
pip_ar_lads = np.loadtxt('pipA_abs/lads-pip-ar-abs.dat', skiprows=1)
pip_n_lads = np.loadtxt('pipA_abs/lads-pip-n-abs.dat', skiprows=1)
if pip_n_lads.ndim == 1:
    pip_n_lads = pip_n_lads[np.newaxis, :]
pip_xe_lads = np.loadtxt('pipA_abs/lads-pip-xe-abs.dat', skiprows=1)
if pip_xe_lads.ndim == 1:
    pip_xe_lads = pip_xe_lads[np.newaxis, :]

#Load the Navon data for bi, c, fe, li, nb
pip_bi_navon = np.loadtxt('pipA_abs/navon-pip-bi-abs.dat', skiprows=1)
if pip_bi_navon.ndim == 1:
    pip_bi_navon = pip_bi_navon[np.newaxis, :]
pip_c_navon = np.loadtxt('pipA_abs/navon-pip-c-abs.dat', skiprows=1)
if pip_c_navon.ndim == 1:
    pip_c_navon = pip_c_navon[np.newaxis, :]
pip_fe_navon = np.loadtxt('pipA_abs/navon-pip-fe-abs.dat', skiprows=1)
if pip_fe_navon.ndim == 1:
    pip_fe_navon = pip_fe_navon[np.newaxis, :]
pip_li_navon = np.loadtxt('pipA_abs/navon-pip-li-abs.dat', skiprows=1)
if pip_li_navon.ndim == 1:
    pip_li_navon = pip_li_navon[np.newaxis, :]
pip_nb_navon = np.loadtxt('pipA_abs/navon-pip-nb-abs.dat', skiprows=1)
if pip_nb_navon.ndim == 1:
    pip_nb_navon = pip_nb_navon[np.newaxis, :]

######################################################
#Also read in the data for pim in the same scheme
pim_al_ashery = np.loadtxt('pimA_abs/ashery-pim-al-abs.dat', skiprows=1)
pim_bi_ashery = np.loadtxt('pimA_abs/ashery-pim-bi-abs.dat', skiprows=1)
pim_c_ashery = np.loadtxt('pimA_abs/ashery-pim-c-abs.dat', skiprows=1)
pim_fe_ashery = np.loadtxt('pimA_abs/ashery-pim-fe-abs.dat', skiprows=1)
pim_li_ashery = np.loadtxt('pimA_abs/ashery-pim-li-abs.dat', skiprows=1)
pim_nb_ashery = np.loadtxt('pimA_abs/ashery-pim-nb-abs.dat', skiprows=1)

#Nakai data for pim
pim_al_nakai = np.loadtxt('pimA_abs/nakai-pim-al-abs.dat', skiprows=1)
pim_au_nakai = np.loadtxt('pimA_abs/nakai-pim-au-abs.dat', skiprows=1)
pim_cu_nakai = np.loadtxt('pimA_abs/nakai-pim-cu-abs.dat', skiprows=1)
pim_sn_nakai = np.loadtxt('pimA_abs/nakai-pim-sn-abs.dat', skiprows=1)
pim_ti_nakai = np.loadtxt('pimA_abs/nakai-pim-ti-abs.dat', skiprows=1)

#Navon data for pim
pim_bi_navon = np.loadtxt('pimA_abs/navon-pim-bi-abs.dat', skiprows=1)
if pim_bi_navon.ndim == 1:
    pim_bi_navon = pim_bi_navon[np.newaxis, :]
pim_c_navon = np.loadtxt('pimA_abs/navon-pim-c-abs.dat', skiprows=1)
if pim_c_navon.ndim == 1:
    pim_c_navon = pim_c_navon[np.newaxis, :]
pim_fe_navon = np.loadtxt('pimA_abs/navon-pim-fe-abs.dat', skiprows=1)
if pim_fe_navon.ndim == 1:
    pim_fe_navon = pim_fe_navon[np.newaxis, :]
```

```python

# Data are organized in 3 columns, separated by 5 spaces: KE in MeV, value of XS in mb, 1 sigma error in XS in mb
#Plotting the data
#Color scale for the different nuclei:
    #The same nucleus from different experiments will have the same color but different marker styles
    #The color will be chosen to be visually distinct for each nucleus
    #The color will change quasicontinuously with increasing atomic mass

#Plotting setup

#Establish a plot color per nucleus. Be creative with matplotlib named colors. Follow a color scheme for increasing A values
colors = {
    "Li": "rebeccapurple",
    "C": "purple",
    "Al": "blue",
    "Bi": "green",
    "Fe": "red",
    "Nb": "brown",
    "Au": "magenta",
    "Cu": "peru",
    "Sn": "orange",
    "Ti": "grey",
    "Ar": "black",
    "N": "cyan",
    "Xe": "pink"
}


fig = plt.figure(figsize=(12,15))

#Ashery
plt.errorbar(pip_al_ashery[:,0], pip_al_ashery[:,1], yerr=pip_al_ashery[:,2], fmt='o--', label='Ashery pi+ Al Absorption', color=colors['Al'])
plt.errorbar(pip_bi_ashery[:,0], pip_bi_ashery[:,1], yerr=pip_bi_ashery[:,2], fmt='o--', label='Ashery pi+ Bi Absorption', color=colors['Bi'])
plt.errorbar(pip_c_ashery[:,0], pip_c_ashery[:,1], yerr=pip_c_ashery[:,2], fmt='o--', label='Ashery pi+ C Absorption', color=colors['C'])
plt.errorbar(pip_fe_ashery[:,0], pip_fe_ashery[:,1], yerr=pip_fe_ashery[:,2], fmt='o--', label='Ashery pi+ Fe Absorption', color=colors['Fe'])
plt.errorbar(pip_li_ashery[:,0], pip_li_ashery[:,1], yerr=pip_li_ashery[:,2], fmt='o--', label='Ashery pi+ Li Absorption', color=colors['Li'])
plt.errorbar(pip_nb_ashery[:,0], pip_nb_ashery[:,1], yerr=pip_nb_ashery[:,2], fmt='o--', label='Ashery pi+ Nb Absorption', color=colors['Nb'])

#Nakai
plt.errorbar(pip_al_nakai[:,0], pip_al_nakai[:,1], yerr=pip_al_nakai[:,2], fmt='s--', label='Nakai pi+ Al Absorption', color=colors['Al'])
plt.errorbar(pip_au_nakai[:,0], pip_au_nakai[:,1], yerr=pip_au_nakai[:,2], fmt='s--', label='Nakai pi+ Au Absorption', color=colors['Au'])
plt.errorbar(pip_cu_nakai[:,0], pip_cu_nakai[:,1], yerr=pip_cu_nakai[:,2], fmt='s--', label='Nakai pi+ Cu Absorption', color=colors['Cu'])
plt.errorbar(pip_sn_nakai[:,0], pip_sn_nakai[:,1], yerr=pip_sn_nakai[:,2], fmt='s--', label='Nakai pi+ Sn Absorption', color=colors['Sn'])
plt.errorbar(pip_ti_nakai[:,0], pip_ti_nakai[:,1], yerr=pip_ti_nakai[:,2], fmt='s--', label='Nakai pi+ Ti Absorption', color=colors['Ti'])

#ProtoDUNE
plt.errorbar(pip_ar_protodune[:,0], pip_ar_protodune[:,1], yerr=pip_ar_protodune[:,2], fmt='X--', label='ProtoDUNE pi+ Ar Absorption', color=colors['Ar'])

#LADS
plt.errorbar(pip_ar_lads[:,0], pip_ar_lads[:,1], yerr=pip_ar_lads[:,2], fmt='^--', label='LADS pi+ Ar Absorption', color=colors['Ar'])
plt.errorbar(pip_n_lads[:,0], pip_n_lads[:,1], yerr=pip_n_lads[:,2], fmt='^--', label='LADS pi+ N Absorption', color=colors['N'])
plt.errorbar(pip_xe_lads[:,0], pip_xe_lads[:,1], yerr=pip_xe_lads[:,2], fmt='^--', label='LADS pi+ Xe Absorption', color = colors['Xe'])

#Navon
plt.errorbar(pip_bi_navon[:,0], pip_bi_navon[:,1], yerr=pip_bi_navon[:,2], fmt='*--', label='Navon pi+ Bi Absorption', color=colors['Bi'])
plt.errorbar(pip_c_navon[:,0], pip_c_navon[:,1], yerr=pip_c_navon[:,2], fmt='*--', label='Navon pi+ C Absorption', color=colors['C'])
plt.errorbar(pip_fe_navon[:,0], pip_fe_navon[:,1], yerr=pip_fe_navon[:,2], fmt='*--', label='Navon pi+ Fe Absorption', color=colors['Fe'])
plt.errorbar(pip_li_navon[:,0], pip_li_navon[:,1], yerr=pip_li_navon[:,2], fmt='*--', label='Navon pi+ Li Absorption', color=colors['Li'])
plt.errorbar(pip_nb_navon[:,0], pip_nb_navon[:,1], yerr=pip_nb_navon[:,2], fmt='*--', label='Navon pi+ Nb Absorption', color=colors['Nb'])

#plots between data points of same nucleus but different experiments could be added here
plt.plot([pip_ar_protodune[0,0], pip_ar_lads[-1,0]], [pip_ar_protodune[0,1], pip_ar_lads[-1,1]], '--', color=colors['Ar'])
plt.plot([pip_bi_ashery[0,0], pip_bi_navon[0,0]], [pip_bi_ashery[0,1], pip_bi_navon[0,1]], '--', color=colors['Bi'])
plt.plot([pip_c_ashery[0,0], pip_c_navon[0,0]], [pip_c_ashery[0,1], pip_c_navon[0,1]], '--', color=colors['C'])
plt.plot([pip_fe_ashery[0,0], pip_fe_navon[0,0]], [pip_fe_ashery[0,1], pip_fe_navon[0,1]], '--', color=colors['Fe'])
plt.plot([pip_li_ashery[0,0], pip_li_navon[0,0]], [pip_li_ashery[0,1], pip_li_navon[0 ,1]], '--', color=colors['Li'])
plt.plot([pip_nb_ashery[0,0], pip_nb_navon[0,0]], [pip_nb_ashery[0,1], pip_nb_navon[0,1]], '--', color=colors['Nb'])

#Label each line on the plot with the nucleus name at the end of the line
textfontsize = 16
plt.text(pip_al_ashery[-1,0]+5, pip_al_ashery[-1,1], "Al", color=colors['Al'],fontsize=textfontsize)
plt.text(pip_bi_ashery[-1,0]+5, pip_bi_ashery[-1,1], "Bi", color=colors['Bi'],fontsize=textfontsize)
plt.text(pip_c_ashery[-1,0]+5, pip_c_ashery[-1,1], 'C', color=colors['C'],fontsize=textfontsize)
plt.text(pip_fe_ashery[-1,0]+5, pip_fe_ashery[-1,1], 'Fe', color=colors['Fe'],fontsize=textfontsize)
plt.text(pip_li_ashery[-1,0]+5, pip_li_ashery[-1,1], "Li", color=colors['Li'],fontsize=textfontsize)
plt.text(pip_nb_ashery[-1,0]+5, pip_nb_ashery[-1,1], 'Nb', color=colors['Nb'],fontsize=textfontsize)
plt.text(pip_au_nakai[-1,0]+5, pip_au_nakai[-1,1], 'Au', color=colors['Au'],fontsize=textfontsize)
plt.text(pip_cu_nakai[-1,0]+5, pip_cu_nakai[-1,1]+50, 'Cu', color=colors['Cu'],fontsize=textfontsize)
plt.text(pip_sn_nakai[-1,0]+5, pip_sn_nakai[-1,1], 'Sn', color=colors['Sn'],fontsize=textfontsize)
plt.text(pip_ti_nakai[-1,0]+5, pip_ti_nakai[-1,1], 'Ti', color=colors['Ti'],fontsize=textfontsize)
plt.text(pip_ar_protodune[-1,0]+5, pip_ar_protodune[-1,1], 'Ar', color=colors['Ar'],fontsize=textfontsize)
#plt.text(pip_ar_lads[-1,0]+5, pip_ar_lads[-1,1], 'Ar', color=colors['Ar'],fontsize=textfontsize)
plt.text(pip_n_lads[-1,0]+5, pip_n_lads[-1,1], 'N', color=colors['N'],fontsize=textfontsize)
plt.text(pip_xe_lads[-1,0]+5, pip_xe_lads[-1,1], 'Xe', color=colors['Xe'],fontsize=textfontsize)

plt.xlabel('Kinetic Energy (MeV)', fontsize=14)
plt.ylabel('Absorption Cross Section (mb)', fontsize=14)
plt.title('Positive Pion Absorption Cross Sections', fontsize=16)
plt.yscale('log')
plt.ylim(20,2500)

#Put the legend off to the side
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='medium')

plt.grid()
plt.savefig('pip_absorption_cross_sections.png', bbox_inches='tight', dpi = fig.dpi)
plt.show()
```
